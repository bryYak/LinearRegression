```python
import numpy as np
from linear_regression import LinearRegression
import matplotlib.pyplot as plt
```

## Creating Training Data

Use _W, _B, _M to define the slope, offset and number of samples for our training data to have. Adjust _noise to add more or less variance to the training data

```python
_W = .3
_B = 50
_M = 500
_noise = 10
```

```python
x = np.linspace(0, 100, _M, dtype=np.float64)
y = np.linspace(_B, _B + 100 * _W, _M, dtype=np.float64)

noise_factor = 0.5

def noise(k):
   return k+(np.random.normal(0, _noise))*noise_factor

x = np.vectorize(noise)(x)
y = np.vectorize(noise)(y)
```

Add Random Noise

```python
plt.scatter(x, y) # For a scatter plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Samples')

plt.show()
```

```python
model = LinearRegression(0.0005, 100000)
x = x.reshape(-1, 1)
model.fit(x, y)
W, B = model.get_weights()
```

```python
x_line = np.linspace(min(x), max(x), 100)
y_line = W * x_line + B
plt.scatter(x, y) # For a scatter plot
plt.plot(x_line, y_line, label = "Regression Line", color="red")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Regression Line')
plt.legend()
plt.show()
```

```python
model = LinearRegression(5.0e-7, 1000)
x_2 = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_2 = np.array([460, 232, 178])
model.fit(x_2, y_2)
w, b = model.get_weights()
pred = w.dot(x_2.T) + b
print(w)
print(pred, y_2)
```
